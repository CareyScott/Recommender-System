# 5.0 Recommender Systems

# 5.1. What are recommender systems?

Recommender systems are an algorithm which can find and suggest appropriate matches to the end user from a particular dataset provided. This can be either from a content based or collaborative approach. These recommendations can range from social media suggestions, movie, and shopping recommendations. Recommender systems are used widely by big corporations such as Google, Facebook, and Twitter to greater understand their user.

# 5.2. Approaches to recommender systems

There are several different approaches upon creating a recommender system. The two which main examples being Content-Based filtering, and collaborative based filtering.

   ## 5.2.1. Content-Based Approach

This method of recommender system content-based filtering will consider an item&#39;s properties, comparing them to other items properties and proposing appropriate suggestions based on the similarities found between the datasets. This form of recommender system does not store any user data or preferences and can be easily adjusted based on the user&#39;s present preferences. This form of Recommender system requires a great amount of data to be able to provide an accurate output to the user.

   ## 5.2.2. Collaborative Approach

The collaborative approach of AI recommends items based on previous ratings and preferences set from other users in the past which are gathered and presented to the end user. Some common examples of this can be seen with Netflix, Amazon and many other widely used applications.

   ### 5.2.2.1. Memory Based

This form of approach uses past event data from an existing memory often being that of likes, clicks and other user provided data. This establishes correlations between two users as they may like or view similar items. It may also consider other users&#39; data. User based recommender systems are used to recommend an item based on similarities between many users. &quot;Customers that are like you based on X, Y or Z&quot;. This can take age, demographic and similar interests into consideration when suggesting items to the end user. Applications such as Netflix use a previous user interests and ratings to suggest related items to prompt to the end user.

 **Example:** 

Take the following into consideration:

 **Shopper A**  and **Shopper B**  are unrelated. They are both rating products online.

 **Shopper A** has rated  **Product X** 5 stars and  **Product Y** with a 5-star rating.  **Shopper B**  rates  **Product Y**  with a 5-star rating.  **Product X**  would therefore be presented to the  **Shopper B**  based on  **Shopper A&#39;s**  similar interest.

   ### 5.2.2.2. Model Based

A model is a question answering system which would be expected to return a correct answer much of the time. A model is built up on data which is gathered from a dataset. A table can be produced presenting the gathered data from the datasets. The data can then be prepared for use once the data us gathered.

The sklearn.feature library is used to import the &#39;TfidfVectorizer&#39;. This is the beginning of the algorithm that recommends to the user. The &#39;tfv\_matrix&#39; is also included. This evaluates the relevance of the words in the dataset and determines their importance which is used in NLP. This matrix enables the data to be filtered by title.



```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vector_game = TfidfVectorizer(min_df = 3,
                         max_features = None,
                         strip_accents = 'unicode',
                         analyzer = 'word',
                         token_pattern = '\w{2,}',
                         ngram_range = (1,3),
                         stop_words = 'english')
```

```python
tfv_matrix=vector_game.fit_transform(games["Title"])
#after recording the video i remembered this applies the parameters to the game inputted
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
```

&#39;sigmoid\_kernal&#39; is originated from Neural Networks. It is an algorithm used for the analysis of the data provided. A sigmoid is known as a hyperbolic tangent which determines an output for each object of the dataset.

```python
sig_game=sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
index = pd.Series(games.index, index=games['Title']).drop_duplicates()
# index = pd.Series(games.index, index=games['Title'])
```

The final stage of the setup of the recommendation system is to create the similarity score function. A new function, &#39;recommend\_game&#39; is created which will be used to run the recommendation. Two variables are set within this function – title and sig\_kern (which is redefined from sig\_game from the previous chunk of code.) &#39;indx&#39; is defined as the index of the game from the dataset (e.g., &quot;Madden NFL 2005&quot; has the index of &#39;8&#39;).

The line &#39;sigmoid\_score = sorted(sigmoid\_score, key = lambda x:x[1], reverse = True)&#39; simply sorts the data in the CSV file.

&#39;sigmoid\_score[1:4]&#39; determines the number of suggested video games presented to the user.

```python
def recommend_game(title, sig_kern=sig_game):
    indx = index[title]

    sigmoid_score = list(enumerate(sig_kern[indx]))
    sigmoid_score = sorted(sigmoid_score, key = lambda x:x[1], reverse = True)
    sigmoid_score = sigmoid_score[1:4]
    game_found = [i[0] for i in sigmoid_score]
    return games.iloc[game_found]
```

A Final Linne of code may be run to input a name of a video game to the system with the expected output of three suggested games based off the algorithm detailed above. This code is represented as "recommend_game('game_title_here')"

```python
recommend_game('Rhythm Heaven')
```

```python
recommend_game('Advance Wars: Days of Ruin')
```
